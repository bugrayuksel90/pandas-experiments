#### loc ve iloc tekrarı

+ <b>loc:</b> Pandas kütüphanesinde veri seçmek için kullanılan çok güçlü bir yöntemdir. Özellikle <i>etiket</i> yani satır ve sütun isimlerini üzerinden nokta atışı veri seçimi yani select işlemi yapılır. "loc" etiket bazlı selection yapar.

+ <b>iloc:</b> Pandas kütüphanesinde veri seçmek için kullanılır. "loc" farkı ise "iloc" <i>pozisyona</i> göre seçim yapar yani satır numarası ve sütun numarası üzerinden işlem yapar. "iloc" integer

```python
#loc ile age ve sex i gösterelim

df.loc[:, ['Age', 'Sex']]
```

```python
# iloc ile BP, Cholesterol,	Na_to_K, Drug gösterelim

df.iloc[:,  [2,3,4,5]]
```

```python
#iloc ile 7. satır ile 143. satırları gösterelim

df.iloc[[7, 143]]
```

```python
#loc ile 55. satırdan 67. satıra kadar olan Age , Na_to_K ve Drug ları gösterelim

df.loc[55:67, ['Age','Na_to_K', 'Drug']]
```

```python
#loc ile ilk 10daki 'Na_to_K' göre yaş, cinsiyet, Na_to_K  ve ilaçları gösterelim

top_ten_na_to_k = df.sort_values(by='Na_to_K', ascending=False).head(10)
top_ten_na_to_k.loc[:, ['Age', 'Sex', 'Na_to_K', 'Drug']]
```

```python
#loc ile en yüksek kadın Na_to_K değerlerini gösterin

top_ten_na_to_k_female = df[df['Sex'] == 'F'].sort_values(by='Na_to_K', ascending=False).head(10)

top_ten_na_to_k_female.loc[: , ['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K', 'Drug']]
```

```python
#iloc ile en yüksek erkek Na_to_K değerlerini gösterin

top_ten_na_to_k_male = df[df['Sex'] == 'M'].sort_values(by='Na_to_K', ascending=False).head(10)

top_ten_na_to_k_male.iloc[:, [0,1,2,3,4]]
```

```python
# Na_to_K 30.000 dan büyük olanları listeleyin

#Path I

(df.query('Na_to_K > 30.000')
    .loc[:, ['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K', 'Drug']]
    .sort_values(by='Na_to_K', ascending=False)
 )


#Path II

# df[
#     df['Na_to_K'] > 30.000
# ][['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K', 'Drug']].sort_values(by='Na_to_K', ascending=False)

#Path III

# df[
#     df['Na_to_K'] > 30.000
# ].loc[:, ['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K', 'Drug']].sort_values(by='Na_to_K', ascending=False)
```

```python
# Kadınlarda 25 yaşının üzerinde olup 50 yaşın altında olan, kollesterolü yüksek olanları sıralayalım (iloc ile)

(df.query('Age > 25 and Age < 50 and Sex == "F" and Cholesterol == "HIGH"')
    .iloc[:, [0,1,2,3,4,5]]
    .sort_values(by='Age', ascending=False)
 )
```

```python
# Erkeklerde X ilacı kullanıp kollesterolü ve tansiyonu normal olanları listeleyelim (Yaş küçükten büyüğe)

(df.query('Drug == "drugX" and Cholesterol == "NORMAL" and BP == "NORMAL" and Sex == "M"')
    .loc[:, ['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K', 'Drug']]
    .sort_values(by='Age', ascending=True)
 )
```

```python
df_fuel = pd.read_csv(filepath_or_buffer='data/FuelConsumption.csv')
df_fuel[:5]
```

```python
df_fuel.shape
```

```python
df_fuel.dtypes
```

```python
# VEHICLECLASS da size içerenleri ve FUELTYPE E olanların 'MAKE', 'VEHICLECLASS', 'ENGINESIZE', 'CYLINDERS', 'TRANSMISSION', 'FUELTYPE' ları listeleyelim

(df_fuel.query('VEHICLECLASS.str.contains("SIZE") and FUELTYPE == "E"')
    .loc[:, ['MAKE', 'VEHICLECLASS', 'ENGINESIZE', 'CYLINDERS', 'TRANSMISSION', 'FUELTYPE']]
    .sort_values(by='CYLINDERS', ascending = False)
 )
```

```python
# Markalara göre karbonemisyon ortalamalarını alalım

df_fuel.groupby('MAKE')[['CO2EMISSIONS']].mean()
```

```python
# Markalara göre FUELCONSUMPTION_CITY	FUELCONSUMPTION_HWY	FUELCONSUMPTION_COMB	FUELCONSUMPTION_COMB_MPG	CO2EMISSIONS ların ortalamalarını ve orta değerlerini listeleyelim

df_fuel.groupby('MAKE')[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']].agg(['mean', 'median'])
```

```python
# markaalakarın modellerine göre FUELCONSUMPTION_COMB CO2EMISSIONS değerlerinin mean ve medianını listeleyelim

df_fuel.groupby(['MAKE','MODEL'])[['FUELCONSUMPTION_COMB','CO2EMISSIONS']].agg(['mean','median'])
```

```python
#bütün markaları görmek için ilk değerlerini listeleyelim

df_fuel.groupby('MAKE').first() # .last() fonksiyonu ile de son değerleri listelenir
```

```python
# Sadece BMW'nin modellerinin mean ve medianlarnı listeleyelim

(df_fuel.query('MAKE == "BMW"')
    .groupby(['MAKE', 'MODEL'])[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
    .agg(['mean', 'median']))
```

```python
# SUV lerin min ve max değerlerini listeleyelim

(df_fuel.query('VEHICLECLASS.str.contains("SUV")')
    .groupby(['MAKE','MODEL','VEHICLECLASS'])[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
    .agg(['min','max']))
```

```python
# benzin tiplerine göre mean ve mediları sıralayalım

df_fuel.groupby('FUELTYPE')[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']].agg(['mean', 'median'])
```

```python
df_fuel.groupby(['MAKE']).groups # groups fonksiyonu index numberlarını gösterir
```

```python
df_fuel.groupby(['MAKE']).groups.keys() # isimlerini döner
```

```python
# Apply fonksiyonu ile  3.5 ile 6.5 motor hacmi arasındaki 'mean','median', 'min', 'max' değerlerini listeleyelim

def min_max_mean_median(avg):
    return avg.agg(['mean','median', 'min', 'max'])



(df_fuel.query('ENGINESIZE >= 3.5 and ENGINESIZE <= 6.5')
    .groupby('ENGINESIZE')[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
    .apply(min_max_mean_median))
```
