```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_products = pd.read_csv(filepath_or_buffer='data/products.csv')
df_products.head()
```

```python
# CategoryID göre ürünlerin stok durumu
result = df_products.groupby('CategoryID')[['UnitsInStock']].sum()
result.sort_values(by='UnitsInStock',  ascending=False)
```

```python
# Hangi tedarikçiden kaç tane ürün alıyoruz
(df_products.groupby('SupplierID')[['ProductID']]
 .count()
 .sort_values(by='ProductID' , ascending=False)
 .rename(
     columns={
         'ProductID' : 'TotalProduct'
     }
 ))
```

```python
df_orders = pd.read_csv(filepath_or_buffer='data/orders.csv')
df_orders.head()
```

```python
df_orders.groupby('ShipVia')[['Freight']].sum().sort_values(by='Freight', ascending=False)
```

```python
# Hangi çalışan kaç tane sipariş gerçekleştirmiş

(df_orders
    .groupby('EmployeeID')[['OrderID']]
    .count()
    .sort_values(by='OrderID', ascending=False)
    .rename(
        columns={
         'OrderID' : 'TotalOrder'
     }))
```

```python
# Hangi müşteri kaç tane sipariş vermiştir

df_grouped = (df_orders
        .groupby('CustomerID')[['OrderID']]
        .count()
        .sort_values(by='OrderID', ascending=False)
        .rename(
            columns={
            'OrderID' : 'TotalOrder'
        }))

#df_grouped['CustomerID'] --> kodu ilgili veri setine "CustomerID" adında bir sütun açtık.
#df_grouped.index --> kodu veri setinin hali hazırda var olan index bilgilerine aldık. Aynı "df_grouped.columns" ta sütun işlemlerini almamız gibi

df_grouped['CustomerID'] = df_grouped.index
#reset_index() --> fonksiyonu ile var olan indexlerin sıfırlayarak 0'dan başlayıp pozitif yönde artan index numarasını verilir.
df_grouped.reset_index(drop=True)
```

```python
# Ürünleri kategorilerine göre gruplayaylım stock mikterı 50 nin altında olanları listeleyelim

#Path I

# (df_products.query('UnitsInStock < 500')
#     .groupby('CategoryID')[['UnitsInStock']]
#     .sum()
#     .sort_values(by='UnitsInStock', ascending=False))

#Path II

(df_products.query('UnitsInStock < 500')
    .groupby('CategoryID')[['UnitsInStock']]
    .agg({'UnitsInStock' : 'sum'})
    .sort_values(by='UnitsInStock', ascending=False))
```

```python
# tedarikçilerden aldığımız stokların ortalaması nedir ve ortalaması 50 den küçük olanlar filtelensin

(df_products.query('UnitsInStock < 50')
    .groupby('SupplierID')[['UnitsInStock']]
    .mean()
    .sort_values(by='SupplierID', ascending=False)
    )
```

```python
df_od = pd.read_csv(filepath_or_buffer='data/order_details.csv')
df_od.head()
```

```python
df_od.shape
```

```python
# Features Crafting
# Projedeki ihitiyaçlara göre var olan sütunlardaki veriler üzerinde aritmatiksel işlemler ya da string manupülasyonları yapılarak yeni özellik yaratılma işlemine denir
# df_od sütununda olmayan "Revenue" sütunu craft edelim. Bunun için UnitPrice *Quantity *(1-Discount) formülünü kullanalım
df_od['Revenue'] = df_od['UnitPrice'] * df_od['Quantity'] * (1-df_od['Discount'])
df_od.head()
```

```python
# Siparişlere göre yeni cirom nedir? 1750 altı 1500 üstüne göre listeleyelim

(df_od
    .groupby('OrderID')[['Revenue']]
    .sum()
    .sort_values(by='Revenue', ascending=False)
    .query('Revenue < 1750 and Revenue > 1500'))
```
