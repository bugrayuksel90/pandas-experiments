## Select

#### Python'da iki tip selector bulumaktadır:

+ <b>loc:</b> Pandas kütüphanesinde veri seçmek için kullanılan çok güçlü bir yöntemdir. Özellikle <i>etiket</i> yani satır ve sütun isimlerini üzerinden nokta atışı veri seçimi yani select işlemi yapılır. "loc" etiket bazlı selection yapar.

+ <b>iloc:</b> Pandas kütüphanesinde veri seçmek için kullanılır. "loc" farkı ise "iloc" <i>pozisyona</i> göre seçim yapar yani satır numarası ve sütun numarası üzerinden işlem yapar. "iloc" integer
+ <hr>

<b>Dilimleme: </b> <i>Hem "loc" hemde "iloc" ile tıpkı python'da ki gibi dilimleme yapılabilinir.</i>

```python
df.head()
```

```python
#loc örneği ile oyuncuların Name, Team ve Pozisyonlarını gösterelim

df.loc[:, ['Name' ,'Team', 'Position' ]]
```

```python
#iloc örneği ile oyuncuların Name, Team ve Pozisyonlarını gösterelim

df.iloc[:, [0,1,3]]
```

```python
#loc örneği ile listedeki ilk 10 oyuncunun Name, Team ve Pozisyonlarını gösterelim
df.loc[1:10, ['Name' ,'Team', 'Position' ]]
```

```python

#listedeki son 10 oyuncunun 'Weight','Salary' lerini gösterelim
df['College'] = df['College'].fillna('?') #College sütunun da NaN olarak gelenleri verinin daha okunabilir olması için '?' imgesine çevrildi
df.loc[df.index[-10:], ['Weight','Salary','College']]
```

```python
#En çok maaş alan ilk 10 oyuncuyu iloc örneği ile 'Name', 'Team', 'Position', 'College', 'Salary' gösterelim
top_ten_player = df.sort_values(by='Salary', ascending=False).head(10) #maaşa göre sıralama yaptırmak istediğiöiz için öncelikle sort_values built_in 
#functionı kullandık ardınladan ilk 10 için head(10)u kullandık
top_ten_player.iloc[:,[0,1,3,7,8]]
```

```python
# Los Angles Lakers da oynayan oyuncuları loc ile bütün satırları gösterelim 
df.loc[df['Team'] == 'Los Angeles Lakers']
```

```python
#En çok kazanan ilk 10 PointGuard oyuncuların 'Name', 'Team', 'Number',	'Position', 'Age', 'Weight', 'Salary' sütularını iloc ile yazdıralım

top_ten_point_guards = df[df['Position'] == 'PG'].sort_values(by='Salary', ascending=False).head(10)
top_ten_point_guards.iloc[:, [0,1,2,3,4,6,8]]
```
